# Label Encoding

```python
# Create a label encoder object
label_encoder = LabelEncoder()

# Fit label encoder on the categorical data and convert categorical labels to numerical values
df['Encoded_URL_Type_obf_Type'] = label_encoder.fit_transform(df["URL_Type_obf_Type"])

# Map "defacement" to 1 and "benign" to 0
df['Encoded_URL_Type_obf_Type'] = df['Encoded_URL_Type_obf_Type'].map({label_encoder.transform(["Defacement"])[0]: 1, label_encoder.transform(["benign"])[0]: 0})

# Print the updated DataFrame
df.drop("URL_Type_obf_Type", axis=1, inplace=True)
df
```

# Scale X between 1 and -1

```python
# Extract the input features (X) to scale
X_unscaled = df_filled.drop('Encoded_URL_Type_obf_Type', axis=1)

# Extract the target variable (y)
y = df_filled['Encoded_URL_Type_obf_Type']

# Create an instance of MinMaxScaler
scaler = MinMaxScaler(feature_range=(-1, 1))

# Select the columns to scale
columns_to_scale = X_unscaled.columns

# Apply Min-Max scaling to the selected columns
scaled_data = scaler.fit_transform(X_unscaled[columns_to_scale])

# Create a new DataFrame with the scaled data
df_scalled = pd.DataFrame(scaled_data, columns=columns_to_scale)

# Add target variable to the scaled DataFrame
df_scalled['Encoded_URL_Type_obf_Type'] = y

df_scalled
```

# Split the data into test, train and validation data

```python
X_train, y_train, X_val, y_val, X_test, y_test = train_valid_test_split(df_scalled, target = 'Encoded_URL_Type_obf_Type', 
                                                                            train_size=0.7, valid_size=0.15, test_size=0.15)
```

# Train the model

```python
from tensorflow.keras import layers, callbacks
# Define model architecture
model = Sequential()
model.add(LSTM(64, input_shape=(X_train.shape[1],1)))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])

early_stopping = callbacks.EarlyStopping(
    min_delta=0.001, # minimium amount of change to count as an improvement
    patience=20, # how many epochs to wait before stopping
    restore_best_weights=True,
)

# Train the model
history = model.fit(X_train, y_train, epochs=200, batch_size=25, validation_data=(X_val, y_val), callbacks=[early_stopping])

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test loss: {loss}, Test accuracy: {accuracy}')
```

```python
#Calculate the confusion matrix
conf_matrix = confusion_matrix(y_true_classes, y_pred_classes)
TN, FP, FN, TP = conf_matrix.ravel()
    
# Print the counts
print("True Positives (TP):", TP)
print("True Negatives (TN):", TN)
print("False Positives (FP):", FP)
print("False Negatives (FN):", FN)

# Define labels for each cell in the confusion matrix
labels = [f"TN\n{conf_matrix[0,0]}", f"FP\n{conf_matrix[0,1]}", f"FN\n{conf_matrix[1,0]}", f"TP\n{conf_matrix[1,1]}"]

# Plot confusion matrix with labels
sns.heatmap(conf_matrix, annot=np.array(labels).reshape(2, 2), fmt="", cmap="Blues")
plt.xlabel("Predicted labels")
plt.ylabel("True labels")
plt.title("Confusion Matrix")
plt.show()
```

```python
# Calculate precision
precision = TP / (TP + FP)

# Calculate recall
recall = TP / (TP + FN)

# Calculate accuracy
accuracy = (TP + TN) / (TP + TN + FP + FN)

# Calculate F1-score
f1_score = 2 * (precision * recall) / (precision + recall)

# Print the evaluation metrics
print("Precision:", precision)
print("Recall:", recall)
print("Accuracy:", accuracy)
print("F1-score:", f1_score)
```

# K-Fold Cross Validation

```python
X = df_scalled.iloc[:, :-1].values
y = df_scalled.iloc[:, -1].values
```

```python
import numpy as np
from sklearn.model_selection import KFold
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam

# Assuming you have already defined X and y as your dataset and labels

# Define model architecture
def build_lstm_model():
    model = Sequential()
    model.add(LSTM(64, input_shape=(X.shape[1], 1)))
    model.add(Dense(1, activation='sigmoid'))
    return model

# Create the model
model = build_lstm_model()
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])

# Set the number of cross-validation folds
num_folds = 5

# Initialize the KFold cross-validator
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)

# Lists to store cross-validation results
accuracy_scores = []
loss_scores = []

# Perform cross-validation
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Reshape X_train and X_test if needed (assuming you have a 3D input shape)
    # X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
    # X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

    # Fit the model on the training data
    model.fit(X_train, y_train, epochs=74, batch_size=25, verbose=1)

    # Evaluate the model on the test data
    loss, accuracy = model.evaluate(X_test, y_test, verbose=1)
    accuracy_scores.append(accuracy)
    loss_scores.append(loss)

# Calculate the average accuracy across all folds
average_accuracy = np.mean(accuracy_scores)

print(f'Average Cross-Validation Accuracy: {average_accuracy}')
```
